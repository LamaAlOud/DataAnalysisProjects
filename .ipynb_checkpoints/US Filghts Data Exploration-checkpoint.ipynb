The day #6 which is "Saturday" have the least cancellation of flight in 2008 (It's weekend so may that why have the least cancellation because the flight may will fully or it is important to make sure the flight goes to its destination for some purposes)

```python
# Cancellation Code
# CancellationCode: reason for cancellation (A = carrier, B = weather, C = NAS, D = security)
CancellationCode_series = df_flight_2008.CancellationCode[df_flight_2008.Cancelled == 1]
dict_CancellationCode_series = dict(CancellationCode_series.value_counts())

CancellationCode = {'A':'carrier', 'B':'weather', 'C':'NAS', 'D':'security'}

plt.figure(figsize=[8, 5])
for key in dict_CancellationCode_series.keys():
    key_value = dict_CancellationCode_series.get(key)
    plt.bar(key, key_value, label=key+' '+CancellationCode.get(key))
    plt.text(key, key_value, str(key_value), ha='center', va='bottom')
plt.legend()
plt.title('Cancellation Reason frequency')
plt.xlabel('Cancellation Code')
plt.ylabel('Count');
```

From above we can generally conclude that the most reason for cancellation because of weather or carrier.

## Bivariate Exploration


```python
# let first have df with only cancelation flights
df_flight_2008_cancellation = df_flight_2008[df_flight_2008.Cancelled == 1]
```

```python
df_flight_2008_cancellation.dropna(axis=1).shape
```

```python
sns.pairplot(df_flight_2008_cancellation[['Month', 'DayOfWeek', 'DayofMonth', 'CRSDepTime', 'CRSArrTime', 'UniqueCarrier', 'FlightNum', 'Distance', 'CancellationCode', 'Origin', 'Dest']])
```

```python
sns.pairplot(df_flight_2008_cancellation.dropna(axis=1))
```

Because the target variable are categorical so we will group by variables to find insights

---

plot for each month which day of year or day of month are the highest day of cancellation

```python
df_month_dayweek_counts = df_flight_2008_cancellation[['Month','DayOfWeek']].groupby(['Month','DayOfWeek'])[['DayOfWeek']].sum()
```

```python
# plot data
fig, ax = plt.subplots(figsize=(15,7))
df_month_dayweek_counts.unstack().plot(ax=ax)
plt.xticks(range(1,13))
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.title('Total sum of days in week "1 (Monday) - 7 (Sunday)" cancellation flights per Month')
plt.ylabel('Sums the Day of Week');
```

From above graph, in months 2(February), 6(June), 7(July), and 12(December) of day 5 for February & December, and day 7 for June & July have peak values, so we conclude the most cnacellation happen in the friday and sunday of February, June, July, and December, but the most cancellation month is February.

```python
df_month_daymonth_counts = df_flight_2008_cancellation[['Month','DayofMonth']].groupby(['Month','DayofMonth'])[['DayofMonth']].sum()
```

```python
# plot data
fig, ax = plt.subplots(figsize=(15,7))
df_month_daymonth_counts.unstack().plot(ax=ax)
plt.xticks(range(1,13))
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5));
```

Because we have 31 day of month the graph above have lots of lines and this will mislead users so in this stage we will not use Day of Month, but we can fighure the 2(Ferbrurary) moth have the most cancellation.

```python
# show each month and its cancellation reason 
df_flight_2008_cancellation[['Month','CancellationCode']].groupby(['Month','CancellationCode'])[['CancellationCode']].count()
```

```python
fig, ax = plt.subplots(figsize=(16,8))
df_flight_2008_cancellation[['Month','CancellationCode']].groupby(['Month','CancellationCode'])[['CancellationCode']].count().unstack().plot.bar(ax=ax)
plt.title('Cancellation Reason for each Month') #  A:carrier, B:weather, C:NAS, D:security
plt.ylabel('Count');
plt.legend(CancellationCode.values());
```

We can see that the most Month have flight cancelation are 2 and 12 in 2008 year, and the cancelled reason is weather, carrier then NAS (National Airspace System)

```python
# number of trips (that cancelled) for each flight number with mean of distances + the cancellation code relate to UniqueCarrier
df_flight_2008_cancellation[['Distance', 'FlightNum', 'UniqueCarrier', 'CancellationCode']].groupby(['UniqueCarrier', 'CancellationCode','FlightNum'])[['Distance']].mean().nlargest(20, 'Distance')
```

```python
# the most 20 number of trips/flight (that cancelled) for each flight number with cancellation code relate to UniqueCarrier
fig, ax = plt.subplots(figsize=(16,8))
df_flight_2008_cancellation[['FlightNum', 'UniqueCarrier', 'CancellationCode']].groupby(['UniqueCarrier', 'CancellationCode', 'FlightNum'])[['FlightNum']].count().nlargest(20, 'FlightNum').unstack().plot.bar(ax=ax)
plt.title('Top 20 number of cancelled flights with UniqueCarrier and CancellationCode(B:weather, C:NAS)') #  A:carrier, B:weather, C:NAS, D:security
plt.ylabel('Flight Count');
```
